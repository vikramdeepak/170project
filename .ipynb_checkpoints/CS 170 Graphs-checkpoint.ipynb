```python
# weight is 0<= l(u,v) < 100
import numpy as np
# k is edge, c is vert
```

```python
# small v <= 30, k = 15, c = 1
class Edge:
        def __init__(self, u, v, w):
            self.u = u
            self.v = v
            self.w = w
            
edges = []
for i in range(29):
    e = Edge(i, (i+1) % 29, np.random.randint(0, 100))
    edges.append(e)
    for j in range(np.random.randint(2, 10)):
        e = Edge(i, np.random.randint(0, 29), np.random.randint(1, 100))
        edges.append(e)
        
f = open("30.in", "a")
f.writelines("29")
print(29)
for e in edges:
    print(e.u, e.v, e.w)
    f.writelines(str(e.u) + " "+ str(e.v) + " "  + str(e.w) + "\n")
f.close()
```
